### COMISIONES GER BRAVO

```python
#intento hacer el mismo filtro que en la celda anterior pero usando el metodo isin

german=df_clean[df_clean.Asegurado.isin(["PEREZ HEREDIA EMANUEL RODRIGO", "PERICHAULT LEANDRO IVAN", "CONCI SOFIA EDITH", 
                                         "BRAVO NATALIA SOLEDAD", "PEREZ ROMERO ENRIQUE SERGIO", "PEREZ HEREDIA FACUNDO SERGIO", 
                                         "CRUZ GABRIEL JUAN", "MORZONE AVILA MATHIAS EZEQUIEL"])]
german.groupby(['Asegurado']).sum('Comision_total')
```

```python
german.info()
```

```python
#suma de las comisiones german
german_filtro=german["Comision_total"].sum()
german_filtro
```

```python
print(com_bruta)
print(costos)
```

```python
#funcion de calculo de comisiones German


def  calcula_g (com_bruta, costos):
    part_german= german_filtro/ com_bruta
    costos_german= part_german*costos
    neto_german= german_filtro - costos_german
    print("La participacion de germán en el total es: ", part_german)
    print("El costo atribuible a Germán del total es: ", costos_german)
    print("El monto de comisiones a transferir a Germán es: ", neto_german)
    return

calcula_g(com_bruta, costos)
```

### COMISIONES STEFI POSCA

```python
#STEFI
# filtro todos sus asegurados del df

stefi=df_clean[df_clean.Asegurado.isin(["POSCA STEFANIA", "POSCA SEBASTIAN ALEJANDRO", "RAFFAELLI EDUARDO MARTIN", 
                                        "ZAMBONI MARTA", "RUFFINI CARLA DANIELA", "DONEMBERG NADIR", "CORDOBA SILVINA ANDREA",
                                        "MOTOS ZAMBONI OCTAVIO"])]
stefi.groupby('Asegurado').sum('Comision_total')
```

```python
#suma de las comisiones stefi
stefi_filtro=stefi["Comision_total"].sum()
stefi_filtro
```

```python

#funcion de calculo de comisiones Stefi
def  calcula_s (com_bruta, costos):
    part_stefi= stefi_filtro/ com_bruta
    costos_stefi= part_stefi*costos
    neto_stefi= stefi_filtro - costos_stefi
    print("La participacion de Stefi en el total es: ", part_stefi)
    print("El costo atribuible a Stefi del total es: ", costos_stefi)
    print("El monto de comisiones a transferir a Stefi es: ", neto_stefi)
    return

calcula_s(com_bruta, costos)
```
